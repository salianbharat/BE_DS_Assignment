so lets find out how many are dead and how many survied along with %

```python
GS.head(20)
```

```python
survived = GS['Survived'].sum()
not_survied = GS['Survived'].count () - survived

#Calculate %

Total_Soul = GS['Survived'].count()

Survived_Percent = (survived / Total_Soul)*100
Not_Survived_Pecent = (not_survied / Total_Soul)*100

print("Total Number of Soul travelling on Titanic", Total_Soul)

print("Total Number of Survived souls are:", survived," & In Terms of %:", np.round(Survived_Percent,))

print("Total Number of dead souls are:", not_survied, "& In Terms of %:", np.round(Not_Survived_Pecent,))
```

**Lets Visualize this Data now.**

```python
import matplotlib.pyplot as plt

# Define labels and values
labels = ['Survived', 'Not Survived']
values = [survived, not_survied]

# Plot the graph
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['green', 'red'])
plt.title('Number of Survived vs Not Survived Passengers')
plt.xlabel('Outcome')
plt.ylabel('Number of Passengers')
plt.show()
```
